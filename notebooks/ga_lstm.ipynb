```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
import tensorflow_addons as tfa
import tqdm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
from datetime import datetime
```

```python
url="https://raw.githubusercontent.com/lmcanavals/data_mining/main/data/"
url+="NSE-TATAGLOBAL.csv"
df = pd.read_csv(url)
train_data = df.iloc[:, 1:2].values
df.head()
```

```python
train_data.shape
```

```python
train_data[:5]
```

```python
df2 = df[-100:]
fig = go.Figure(data=[go.Candlestick(x=df2['Date'],
                open=df2['Open'],
                high=df2['High'],
                low=df2['Low'],
                close=df2['Close'])])
fig.show()
```

```python
sc = MinMaxScaler(feature_range=(0, 0.999999))
train_data = sc.fit_transform(train_data)
```

```python
x_train = []
y_train = []
for i in range(60, 2035):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train = x_train.reshape(*x_train.shape, 1)
```

```python
model = Sequential([LSTM(50, input_shape=(x_train.shape[1], 1), return_sequences=True),
                    Dropout(0.2),
                    LSTM(units = 50, return_sequences = True),
                    Dropout(0.2),
                    LSTM(units = 50, return_sequences = True),
                    Dropout(0.2),
                    LSTM(units = 50),
                    Dropout(0.2),
                    Dense(units = 1)])
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
tqdm_callback = tfa.callbacks.TQDMProgressBar()
```

```python
history = model.fit(x_train, y_train, epochs=100, batch_size=32, verbose=0,
                    callbacks=[tqdm_callback])
```
